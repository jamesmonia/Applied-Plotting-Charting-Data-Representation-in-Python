```python
plt.figure()
y = np.random.normal(loc=0.0, scale=1.0, size=10000)
x = np.random.normal(size=10000) #same as above? diff values since its random
plt.scatter(x,y)
```

```python
plt.figure()
gspec = gridspec.GridSpec(3,3)
#creates 3x3 subplots - meaning gspec can be cast as a list of 9 subplots
#down then right
#think in terms of a 3x3 grid - hence it's a gridspec

top = plt.subplot(gspec[0,1:])
# 0 moves from the top-left, 1: moves from the left
side = plt.subplot(gspec[1:,0])
# 1: moves from the top-left, 0 from the left
lower_right = plt.subplot(gspec[1:,1:])
# 1: moves from the top, 1: moves from the left
#since it's a 3x3, [1:,1:] will be positions 1 and 2 for x and 1 and 2 for y (starting at 0 like a normal list)
#treat each one like an ax you can plot onto
#assigning a variable to the ax still shows it on the fig anyway

lower_right.scatter(x,y,alpha=0.1,s=10)

top.hist(x,bins=100,normed=True)
side.hist(y,bins=100,orientation='horizontal',normed=True)
side.invert_xaxis()
```

```python
#animations
import matplotlib.animation as animation
```

```python
n = 200
x = np.random.randn(n)
```

```python
def update(curr):
    if curr == n:
        a.event_source.stop()
    plt.cla() #clear the current axes
    bins = np.arange(-4,4,0.5)
    plt.hist(x[:curr],bins=bins)
    plt.axis([-4,4,0,30])
    
    plt.gca().set_title('Sampling the Norm Dist')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate(f'n= {curr}',[3,27])
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig,update,interval=50) #current figure, the function we're calling, time interval in ms
```

```python
plt.figure()
data = np.random.rand(10)
plt.plot(data)

def onclick(event):
    plt.cla() #clear current axis
    plt.plot(data)
    plt.gca().set_title(f'Event at pixels {event.x},{event.y} and data {event.xdata},{event.ydata}')
                        
plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```

```python
from random import shuffle
```

```python
origins = ['China', 'Brazil', 'India', 'USA', 'Canada', 'UK', 'Germany', 'Iraq', 'Chile', 'Mexico']
shuffle(origins)

df = pd.DataFrame({'height': np.random.rand(10),
                   'weight': np.random.rand(10),
                   'origin': origins})
df
```

```python
plt.figure()
plt.scatter(df['height'],df['weight'],picker=5) #picker is the number of pixels away a click can be and still hit the object
plt.gca().set_ylabel('Weight')
plt.gca().set_xlabel('Height')
```

```python
def onpick(event):
    #print(event)
    origin = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title(f'Selected item came from {origin}')
    
plt.gcf().canvas.mpl_connect('pick_event', onpick)
#adds a pick_event listener to the canvas with the given function. 
#the event that gets passed in has certain attributes which, in this case, happen to line up with the index of our df
#we're selecting the origin column to display 
```
